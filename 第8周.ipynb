```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

df2 = pd.DataFrame(np.random.randn(100, 4), columns=list("ABCD"))
X = df2['A']
y = df2['B']
# Add a constant to the model (for the intercept term)
X = sm.add_constant(X)
# Fit the regression model
model = sm.OLS(y, X).fit()
X
# Display the regression results
print(model.summary())
```

```python

#複回歸
import pandas as pd
import statsmodels.api as sm
import numpy as np

df2 = pd.DataFrame(np.random.randn(100, 4), columns=list("ABCD"))
y = df2['B']  # 應變數
X = df2[['A', 'C', 'D']]  # 自變數
X = sm.add_constant(X) # Add a constant to the model (for the intercept term)
model = sm.OLS(y, X).fit()
# Fit the regression model
print(model.summary())
```
